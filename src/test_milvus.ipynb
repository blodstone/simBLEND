```python
from pymilvus import MilvusClient
import torch
```

```python

client = MilvusClient("milvus_demo.db")
```

```python
if client.has_collection(collection_name="demo_collection"):
    client.drop_collection(collection_name="demo_collection")
client.create_collection(
    collection_name="demo_collection",
    dimension=768,  # The vectors we will use in this demo has 768 dimensions
)
```

```python
# Convert the tensor to a list of dicts as required by MilvusClient
vectors = torch.randn(100, 768).tolist()
data = []
for i in range(100):
    data.append({"vector": vectors[i], "id": i, "text": f"sample text {i}"})
res = client.insert(collection_name="demo_collection", data=data)
```

```python
target_vector = torch.randn(768)
```

```python

res = client.search(collection_name="demo_collection",
                    search_params={"metric_type": "COSINE"}, 
                    anns_field="vector", 
                    data=[target_vector.tolist()],
                    limit=3,output_fields=["id", "text", "vector"])
```

```python
for hits in res:
    for hit in hits:
        print(hit)
```
